```python
## Euler 
def euler_method(y0, t0, h, n, fty):
	expr = parser.expr(fty).compile()

	for i in range(0, n + 1):        
		arr_t.append(t0)
		arr_y.append(y0)

        # yn+1 = yn + h*f(tn, yn)
		y0 += h*do_eval(t0, y0, expr)
		t0 += h

	return t0
```

```python
## Euler Inverso
def backward_euler_method(y0, t0, h, n, fty):
	expr = parser.expr(fty).compile()

	for i in range(0, n + 1):        
		arr_t.append(t0)
		arr_y.append(y0)
        
		# yn+1 = yn + h*f(tn, yn)
		y_euler = y0 + h*do_eval(t0, y0, expr) # descobre yn+1
        
		t0 += h
        
		# yn+1 = yn + h*f(tn+1, yn+1)
		y0 += h*do_eval(t0, y_euler, expr)
	
	return t0
```

```python
## Euler Melhorado
def best_euler_method(y0, t0, h, n, fty):
	expr = parser.expr(fty).compile()

	for i in range(0, n + 1):        
		arr_t.append(t0)
		arr_y.append(y0)
        
		# f(tn, yn)
		f_tn_yn = do_eval(t0, y0, expr)
		
		# yn+1 = yn + h*f(tn, yn)
		y_euler = y0 + h*f_tn_yn # descobre yn+1
		
		t0 += h
		
		# f(tn+1, yn+1)
		f_tn1_yn1 = do_eval(t0, y_euler, expr)
		
		# yn+1 = yn + (h/2)*(f(tn, yn) + f(tn+1, yn+1))
		y0 += (h/2)*(f_tn_yn + f_tn1_yn1)

	return t0
```

```python
## Runge Kutta
def runge_kutta_method(y0, t0, h, n, fty):
	expr = parser.expr(fty).compile()

	for i in range(0, n + 1):        
		arr_t.append(t0)
		arr_y.append(y0)
        
		### Kn1 ###
		# f(tn, yn)
		kn1 = do_eval(t0, y0, expr)

		### Kn2 ###
		# f(tn + h/2, yn + (h/2)*kn1)
		kn2 = do_eval(t0 + h/2, y0 + (h/2)*kn1, expr)

		### Kn3 ###
		# f(tn + h/2, yn + (h/2)*kn2)
		kn3 = do_eval(t0 + h/2, y0 + (h/2)*kn2, expr)

		### Kn4 ###
		# f(tn + h, yn + h*kn3)
		kn4 = do_eval(t0 + h, y0 + h*kn3, expr)
		
		# yn+1 = yn + (h/6)*(kn1 + kn2 + kn3 + kn4)
		y0 += (h/6)*(kn1 + 2*kn2 + 2*kn3 + kn4)
		t0 += h

	return t0
```

```python
## Runge Kutta
def runge_kutta_method(y0, t0, h, n, fty):
	expr = parser.expr(fty).compile()

	for i in range(0, n + 1):        
		arr_t.append(t0)
		arr_y.append(y0)
        
		### Kn1 ###
		# f(tn, yn)
		kn1 = do_eval(t0, y0, expr)

		### Kn2 ###
		# f(tn + h/2, yn + (h/2)*kn1)
		kn2 = do_eval(t0 + h/2, y0 + (h/2)*kn1, expr)

		### Kn3 ###
		# f(tn + h/2, yn + (h/2)*kn2)
		kn3 = do_eval(t0 + h/2, y0 + (h/2)*kn2, expr)

		### Kn4 ###
		# f(tn + h, yn + h*kn3)
		kn4 = do_eval(t0 + h, y0 + h*kn3, expr)
		
		# yn+1 = yn + (h/6)*(kn1 + kn2 + kn3 + kn4)
		y0 += (h/6)*(kn1 + 2*kn2 + 2*kn3 + kn4)
		t0 += h

	return t0
```

### Primeira QuestÃ£o

#### Letra a
Faremos entÃ£o a deduÃ§Ã£o do primeiro problema tomando as seguintes variÃ¡veis 
 
$Q$ - concentraÃ§Ã£o  
$Z$ - vazÃ£o  
$V$ - volume  
$S$ - quantidade de sal 
 

---
Para determinar a concentraÃ§Ã£o $Q(t)$ em um dado momento qualquer, faremos
\begin{equation}
    \label{eq_concentracao}
    Q(t) = \frac{S(t)}{V(t)}
\end{equation}

onde:  

\begin{equation}
    \label{eq_volume}
    V(t) = V_{0} + (Z_{in} - Z_{out})t
\end{equation}

---
Sabemos que a variaÃ§Ã£o da quantidade de sal $dS(t)$ Ã© dada pela diferenÃ§a entre a entrada e a saÃ­da de sal em um dado instante t, isto Ã©

\begin{equation*}
    \label{eq_var_qtd_sal}
    dS(t) = S_{in} - S_{out}
\end{equation*}


\begin{equation*}
    \label{eq_var_qtd_sal_2}
    \frac{dS(t)}{dt} = \frac{dS_{in}}{dt} - \frac{dS_{out}}{dt}
\end{equation*}


\begin{equation*}
    \label{eq_var_qtd_sal_3}
    \frac{dS(t)}{dt} = \frac{d(Q_{in}(t) \cdot V_{in}(t))}{dt} - \frac{d(Q_{out}(t) \cdot V_{out}(t))}{dt}
\end{equation*}

\begin{equation*}
    \label{eq_var_qtd_sal_4}
    S'(t) = (Q'_{in}(t) \cdot V_{in}(t) + Q_{in}(t) \cdot V'_{in}(t)) - (Q'_{out}(t) \cdot V_{out}(t) + Q_{out}(t) \cdot V'_{out}(t))
\end{equation*}


Assumindo que a Ã¡gua que entra no tanque Ã© pura, a concentraÃ§Ã£o de entrada $Q_{in}$ Ã© zero, entÃ£o

\begin{equation*}
    \label{eq_var_qtd_sal_5}
    S'(t) = - (Q'_{out}(t) \cdot V_{out}(t) + Q_{out}(t) \cdot V'_{out}(t))
\end{equation*}

Uma vez que a massa de Ã¡gua que sai em um instante estacionÃ¡rio Ã© nula, podemos assumir que o termo $Q'_{out}(t) \cdot V_{out}(t) = 0$, que nos deixa com o seguinte resultado
\begin{equation*}
    S'(t) = - Q_{out}(t) \cdot V'_{out}(t)
\end{equation*}

Contudo, sabemos que a vazÃ£o Ã© dada por meio de:
\begin{equation}
    \label{eq_vazao}
    Z(t) = V'(t)
\end{equation}

Assim, assumimos por meio de \eqref{eq_vazao} que $S'(t)$ Ã© dada pela concentraÃ§Ã£o de sal no tanque $S(t)$ multiplicado pela vazÃ£o de saÃ­da do tanque $V_{out}$

\begin{equation}
    \label{eq_var_qtd_sal_6}
    S'(t) = - Q(t) \cdot Z_{out}
\end{equation}

Substituindo \eqref{eq_concentracao} e \eqref{eq_volume} em \eqref{eq_var_qtd_sal_6}, teremos:

\begin{equation}
    \label{eq_var_qtd_sal_7}
    S'(t) = - \frac{S(t)}{V(t)} \cdot Z_{out}
\end{equation}

Assim, nossa EDO serÃ¡:

\begin{equation}
    \label{edo1}
    S'(t) + \frac{Z_{out}}{V_{0} + (Z_{in} - Z_{out})t} \cdot S(t) = 0
\end{equation}

Com a substituiÃ§Ã£o dos valores dados na questÃ£o em \eqref{edo1}, teremos:

\begin{equation}
    \label{edo_com_valores}
    S'(t) + \frac{4}{100+2t} \cdot S(t) = 0
\end{equation}

---
Em posse dessa equaÃ§Ã£o diferencial podemos realizar os procedimentos para determinar qual sua soluÃ§Ã£o exata para a quantidade de sal no tanque $S(t)$ em um dado instante $t$. Para isto utilizaremo o mÃ©todos dos fatores integrantes para esta equaÃ§Ã£o.  
O mÃ©todo de fatores integrantes consiste na seguinte hipÃ³tese:  
Seja uma equaÃ§Ã£o diferencial dada por
\begin{equation}
    \label{eq_fatores_integrantes}
    y'(t) + P(t) \cdot y(t) = Q(t)
\end{equation}

Multiplicamos entÃ£o a equaÃ§Ã£o por um fator $\mu (t)$:
\begin{equation}
    \label{eq_fatores_integrantes_1}
    \mu(t) \cdot y'(t) + \mu(t) \cdot P(t) \cdot y(t) = \mu(t) \cdot Q(t)
\end{equation}

DaÃ­ podemos assumir sem perda de generalidade que a funÃ§Ã£o arbitrÃ¡ria $\mu (t)$ Ã© descrita conforme:
\begin{equation}
    \label{eq_fatores_integrantes_2}
    \mu'(t) = mu(t) \cdot P(t)
\end{equation}

E substituindo \eqref{eq_fatores_integrantes_2} em \eqref{eq_fatores_integrantes_1}, que nos deixa com uma expressÃ£o da forma:
\begin{equation*}
    \label{eq_fatores_integrantes_3}
    \mu(t) \cdot y'(t) + \mu'(t) \cdot y(t) = \mu(t) \cdot Q(t)
\end{equation*}

\begin{equation*}
    \label{eq_fatores_integrantes_4}
    \frac{d\mu(t) y(t)}{dt} = \mu(t) \cdot Q(t)
\end{equation*}

Segundo a Lei de Resfriamento de Newton,

\begin{equation}
    \label{eq_newton}

\end{equation}

Onde,  
$T$ - temperatura do corpo no instante t  
$T_{a}$ - temperatura constante do ambiente  
$t$ - tempo  
$k$ - constante  

Resolvendo a EDO, teremos:
\begin{equation*}
    \label{eq_newton_1}
    \frac{dT}{dt} = -k \cdot (T - T_{a})
\end{equation*}

\begin{equation*}
    \label{eq_newton_2}
    \frac{1}{T - T_{a}} \cdot dT = -k \cdot dt
\end{equation*}

Integrando os dois lados

\begin{equation*}
    \label{eq_newton_3}
    \int\frac{1}{T - T_{a}} \cdot dT = -\int k \cdot dt
\end{equation*}


\begin{equation*}
    \label{eq_newton_4}
    \ln(T - T_{a}) = - kt + C_{2} - C_{1}
\end{equation*}

Para tirar o $\ln(T - T_{a})$, faremos

\begin{equation*}
    \label{eq_newton_5}

\end{equation*}

Logo,

\begin{equation*}
    T - T_{a} = e^{-kt} \cdot e^{c}
\end{equation*}

\begin{equation}
